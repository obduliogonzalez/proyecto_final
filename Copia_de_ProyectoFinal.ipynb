```python
#Sustituyendo los valores NA en las columnas Object basado en probabilidades

import pandas as pd
import numpy as np

# Parámetros
chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'

# Función para sustituir valores con base en probabilidades
def replace_with_probabilities(chunk):
    # Identificar columnas de tipo 'object'
    object_columns = chunk.select_dtypes(include=['object']).columns

    for col in object_columns:
        # Calcular distribución de probabilidad de los valores únicos
        value_counts = chunk[col].value_counts(normalize=True)
        unique_values = value_counts.index
        probabilities = value_counts.values

        # Generar nuevos valores basados en las probabilidades
        chunk[col] = np.random.choice(unique_values, size=len(chunk), p=probabilities)

    return chunk

# Procesar archivo en chunks
processed_chunks = []

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    print(f"Procesando chunk {i+1}...")

    # Reemplazar valores en las columnas tipo 'object'
    processed_chunk = replace_with_probabilities(chunk)

    # Almacenar o procesar más el chunk si es necesario
    processed_chunks.append(processed_chunk)

# Combinar todos los chunks si se desea procesar el dataset completo después
final_data = pd.concat(processed_chunks)

# Mostrar algunas filas del resultado
print(final_data.head())


```

```python
#Sustituyendo los valores NA en las columnas que no son Object basado en probabilidades

import pandas as pd
import numpy as np

# Parámetros
chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'

# Función para sustituir valores en columnas numéricas
def replace_numeric_with_random(chunk):
    # Identificar columnas que no son de tipo 'object'
    numeric_columns = chunk.select_dtypes(exclude=['object']).columns

    for col in numeric_columns:
        # Calcular estadísticas básicas
        col_mean = chunk[col].mean()
        col_std = chunk[col].std()

        # Sustituir valores con nuevos generados aleatoriamente
        # Basado en una distribución normal con la misma media y desviación estándar
        chunk[col] = np.random.normal(loc=col_mean, scale=col_std, size=len(chunk))

    return chunk

# Procesar archivo en chunks
processed_chunks = []

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    print(f"Procesando chunk {i+1}...")

    # Reemplazar valores en las columnas numéricas
    processed_chunk = replace_numeric_with_random(chunk)

    # Almacenar o procesar más el chunk si es necesario
    processed_chunks.append(processed_chunk)

# Combinar todos los chunks si se desea procesar el dataset completo después
final_data = pd.concat(processed_chunks)

# Mostrar algunas filas del resultado
print(final_data.head())
```

Ciudades de destinos más frecuente

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros
chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'

# Crear un DataFrame vacío para almacenar los datos procesados
aggregated_data = pd.DataFrame()

# Procesar el archivo por chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Seleccionar la columna de destino
    relevant_data = chunk[['DEST_CITY_NAME']]

    # Agrupar por ciudad de destino y contar las ocurrencias
    grouped_data = relevant_data.groupby('DEST_CITY_NAME').size().reset_index(name='Count')

    # Acumular los datos procesados
    aggregated_data = pd.concat([aggregated_data, grouped_data])

# Volver a agrupar después de procesar todos los chunks
final_data = aggregated_data.groupby('DEST_CITY_NAME')['Count'].sum().reset_index()

# Seleccionar los 15 destinos con más vuelos
top_15_destinations = final_data.nlargest(15, 'Count')

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(data=top_15_destinations, x='DEST_CITY_NAME', y='Count', palette='viridis')

# Configuración del gráfico
plt.title('Top 15 Destinos Más Frecuentes', fontsize=16)
plt.xlabel('Ciudad de Destino', fontsize=14)
plt.ylabel('Cantidad de Vuelos', fontsize=14)
plt.xticks(rotation=90, fontsize=12)  # Rotar las etiquetas de las ciudades
plt.yticks(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

Gráfico de barras en donde se visualiza la cantidad de vuelos por mes



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros
chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'

# Crear un DataFrame vacío para almacenar los datos procesados
aggregated_data = pd.DataFrame()

# Procesar el archivo por chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Seleccionar las columnas necesarias
    relevant_data = chunk[['MONTH', 'DEST_CITY_NAME']]

    # Agrupar por mes y ciudad de destino, y contar los destinos únicos
    grouped_data = relevant_data.groupby(['MONTH', 'DEST_CITY_NAME']).size().reset_index(name='Count')

    # Agrupar por mes para contar la cantidad de destinos únicos
    monthly_dest_count = grouped_data.groupby('MONTH')['DEST_CITY_NAME'].nunique().reset_index(name='UniqueDestCount')

    # Acumular los datos procesados
    aggregated_data = pd.concat([aggregated_data, monthly_dest_count])

# Volver a agrupar después de procesar todos los chunks
final_data = aggregated_data.groupby('MONTH')['UniqueDestCount'].sum().reset_index()

# Seleccionar los meses con mayor cantidad de destinos
top_months = final_data.nlargest(12, 'UniqueDestCount')

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(data=top_months, x='MONTH', y='UniqueDestCount', palette='viridis')

# Configuración del gráfico
plt.title('Mayor Cantidad de Vuelos', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Cantidad de vuelos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'


aggregated_data = pd.DataFrame()
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    relevant_data = chunk[['DEP_DELAY', 'CANCELLED', 'AIR_TIME', 'WEATHER_DELAY']]

    # Acumular los datos procesados
    aggregated_data = pd.concat([aggregated_data, relevant_data])

# Calcular la matriz de correlación
correlation_matrix = aggregated_data.corr()

# Crear el mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Configuración del gráfico
plt.title('Mapa de Calor de Correlaciones entre Variables', fontsize=16)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Parámetros
chunk_size = 100000  # Tamaño del chunk
#file_path = '/content/drive/MyDrive/final_data.csv'
file_path = '/content/drive/MyDrive/Colab Notebooks/final_data.csv'


# Crear un DataFrame vacío para almacenar los datos procesados
sample_data = pd.DataFrame()

# Procesar el archivo por chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Seleccionar las columnas relevantes
    relevant_data = chunk[['MONTH', 'AIRLINE_ID', 'DEST_AIRPORT_ID', 'DEST_CITY_MARKET_ID',
                           'DEP_TIME', 'TAXI_IN', 'TAXI_OUT', 'CANCELLED', 'DISTANCE']]

    # Tomar una muestra aleatoria del chunk para reducir el tamaño
    sampled_chunk = relevant_data.sample(frac=0.01, random_state=42)  # 1% del chunk
    sample_data = pd.concat([sample_data, sampled_chunk])

    # Limitar el tamaño total del sample
    if len(sample_data) > 10000:  # Máximo de 10,000 filas en total
        break

# Crear el pairplot
sns.pairplot(sample_data, diag_kind="kde", plot_kws={'alpha': 0.5}, corner=True )

# Configuración del gráfico
plt.suptitle("Pairplot de Variables Seleccionadas", y=1.02, fontsize=16)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
